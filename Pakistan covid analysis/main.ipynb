```python
import pandas as pd
from matplotlib import pyplot as plt
import datetime as dt
import seaborn as sns

df = pd.read_csv('data/combined_report.csv')
df.head(20)
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df.dropna(subset='Province_State', inplace= True)
df.head(10)
```

```python
df.value_counts(subset='Province_State').index
```

```python
df.Last_Update = pd.to_datetime(df['Last_Update']).dt.date
df['Last_Update'] = pd.to_datetime(df['Last_Update'])
df.head(10)
```

```python
fig = plt.figure(figsize = (14,10))
plt.grid()
graf_with_time = sns.lineplot(data=df[df['Province_State'].isin(list(df.value_counts(subset='Province_State').index))], x = 'Last_Update', y= 'Confirmed', hue='Province_State')
```

```python
mask = (df['Last_Update'] > '2021-05-01')
fig = plt.figure(figsize = (14,10))
plt.grid()
graf_with_time = sns.lineplot(data=df[df['Province_State'].isin(list(df.value_counts(subset='Province_State').index))].loc[mask], x ='Last_Update', y= 'Confirmed', hue='Province_State')
```

```python
totel_per_state = pd.DataFrame()

totel_per_state['Confirmed'] = df.groupby(['Province_State'])['Confirmed'].transform('sum')
totel_per_state['Deaths'] = df.groupby(['Province_State'])['Deaths'].transform('sum')
totel_per_state['Recovered'] = df.groupby(['Province_State'])['Recovered'].transform('sum')
totel_per_state['Active'] = df.groupby(['Province_State'])['Active'].transform('sum')

totel_per_state.drop_duplicates(inplace =True)
totel_per_state.index = list(df.value_counts(subset='Province_State').index)
totel_per_state
```

```python
piller_comparison = plt.figure(figsize = (14,10))
plt.grid()
piller_comparison = plt.title("Confirmed covid cases Comparison", size=25)
piller_comparison = plt.bar(totel_per_state.index, totel_per_state['Confirmed'])
piller_comparison = plt.bar(totel_per_state.index, totel_per_state['Recovered'])
piller_comparison = plt.bar(totel_per_state.index, totel_per_state['Deaths'])
piller_comparison
```
