## web scraping first project 
### Collecting product from draze.pk

```python
# importing 
import requests
from bs4 import BeautifulSoup
import json

for_search = 'books' # product type to search on Draze
from_page = 2 # page on Draze

def soup_content(search, page_no):
    URL = f'https://www.daraz.pk/catalog/?page={page_no}&q={search}'

    respo = requests.get(URL) # doing request fro URL assining responce to respo
    soup = BeautifulSoup(respo.content, 'html.parser') # using BeautifulSoup to make it useable
    return soup
soup = soup_content(for_search, from_page) # soup is content can be use

```

```python

sope_script3 = str(soup.findAll('script')[3]) # converting sope third script to str

itemlists = sope_script3[sope_script3.find('listItems')+len('listItems:[')+1:sope_script3.find('sortItems')] # faching only products as itemlist
itemlist = itemlists.replace('"', '').split(',') # removing and replacing 

prod = []
products = []

for i in itemlist:
    if 'name:' in i:
        name = i.replace('{name:', '')
        prod.append(name)
    if 'productUrl' in i and 'www.daraz.pk' in i:
        proURL = i.replace('productUrl:', '')
        prod.append(proURL)
    if 'ratingScore' in i:
        rating = i.replace('ratingScore:', '')
        prod.append(rating)
    if 'current_price' in i:
        price = i.replace('current_price:', '')
        prod.append(int(float(price)))
    if len(prod)==4:
        products.append(prod)
        prod = []
import pandas as pd

df = pd.DataFrame(products, columns=['Name', 'Price', 'Link', 'Rating'])
df

    
```
