### Collecting product from draze.pk

```python
# importing 
import requests
from bs4 import BeautifulSoup
import json

for_search = 'books' # product type to search on Draze

URL = f"https://www.daraz.pk/catalog/?q={for_search}" # URL with search

respo = requests.get(URL) # doing request fro URL assining responce to respo

soup = BeautifulSoup(respo.content, 'html.parser') # using BeautifulSoup to make it useable
```

```python

sope_script3 = str(soup.findAll('script')[3]) # converting sope third script to str

itemlists = sope_script3[sope_script3.find('listItems')-1:sope_script3.find('sortItems')] # faching only products as itemlist
itemlist = itemlists.replace('"', '').replace('listItems:[', '').split(',') # removing and replacing 

prod = []
products = []

for i in itemlist:
    if 'name:' in i:
        name = i.replace('{name:', '')
        prod.append(name)
    if 'productUrl' in i and 'www.daraz.pk' in i:
        proURL = i.replace('productUrl:', '')
        prod.append(proURL)
    if 'ratingScore' in i:
        rating = i.replace('ratingScore:', '')
        prod.append(rating)
    if 'current_price' in i:
        price = i.replace('current_price:', '')
        prod.append(price)
    # if 'image' in i:
    #     image = i.replace('image:', '')
    #     prod.append(image)
    if len(prod)==4:
        products.append(prod)
        # print(prod)
        prod = []
    # print(name, proURL, rating, price)
# products
import pandas as pd

df = pd.DataFrame(products, columns=['Name', 'Price', 'Link', 'Rating'])
# df.Price.astype('int64')

df.Price.astypes = 'int64'
    
```
