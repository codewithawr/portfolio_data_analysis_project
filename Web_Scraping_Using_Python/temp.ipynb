```python
#This will not run on online IDE
import requests
from bs4 import BeautifulSoup
import json

for_search = 'house'

URL = f"https://www.daraz.pk/catalog/?q={for_search}"
r = requests.get(URL)

soup = BeautifulSoup(r.content, 'html.parser') # If this line causes an error, run 'pip install html5lib' or install html5lib
# print(soup.prettify())
# json.loads(u)

itemlists = str(soup.findAll('script')[3])[str(soup.findAll('script')[3]).find('listItems')-1:str(soup.findAll('script')[3]).find('sortItems')]
```

```python
itemlist = itemlists.replace('"', '').replace('listItems:[', '').split(',')

count = 0
prod = []
products = []

for i in itemlist:
    if 'name:' in i:
        name = i.replace('{name:', '')
        prod.append(name)
    if 'productUrl' in i and 'www.daraz.pk' in i:
        proURL = i.replace('productUrl:', '')
        prod.append(proURL)
    if 'ratingScore' in i:
        rating = i.replace('ratingScore:', '')
        prod.append(rating)
    if 'current_price' in i:
        price = i.replace('current_price:', '')
        prod.append(price)
    count += 1
    if len(prod)==4:
        products.append(prod)
        # print(prod)
        prod = []
    # print(name, proURL, rating, price)
# products
import pandas as pd

df = pd.DataFrame(products)
df
    
```
