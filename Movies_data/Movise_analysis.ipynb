```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# setting's

# for not to convert velues to e to the power
pd.set_option('display.float_format', '{:.2f}'.format)

# size of plots
plt.rcParams['figure.figsize'] = [12, 7]
```

## Importing Data

```python
df = pd.read_csv('Data/movies.csv')
df.head()
```

## checking null velues avarage

```python
null_velues_column = {}

for col in df.columns:
    null_velues_column[col] = np.mean(df[col].isnull())

null_velues_column = pd.DataFrame([null_velues_column])
null_velues_column
```

```python
df.head(3)
```

## converting corect dtypes

```python
# df.votes = df.votes.values.astype(np.int64)
# df.gross = df.gross.values.astype(np.int64)
df.runtime = df.runtime.values.astype(np.int64)
df.budget = df.budget.values.astype(np.int64)

df.dtypes
```

## delling with dublicates

```python
df.drop_duplicates(inplace=True)
df.head()
```

### odering over data

```python
df.describe('')
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
df.head(3)
```

## finding realation is year affecting to gross

```python
sns.regplot(df.year, df.gross, scatter_kws={'color': 'blue'}, line_kws={'color': "red"})
plt.title("Gross by Year")
plt.xlabel("Year")
plt.ylabel("Gross")
```

## gross groupby score

```python
avrage_grose_per_score = pd.DataFrame()
avrage_grose_per_score_temp = pd.DataFrame()
avrage_grose_per_score_temp['score'] = df['score'].round(0)
avrage_grose_per_score_temp['gross'] = df['gross']
avrage_grose_per_score = avrage_grose_per_score_temp.groupby('score')['gross'].mean()
```

```python
sns.barplot(avrage_grose_per_score.index, avrage_grose_per_score)
plt.title("Gross by Score")
plt.xlabel("Score")
plt.ylabel("Gross")
```

## finding which genra hase most gross

```python
avrage_grose_per_genre = pd.DataFrame()
avrage_grose_per_genre_temp = pd.DataFrame()
avrage_grose_per_genre_temp['genre'] = df['genre']
avrage_grose_per_genre_temp['gross'] = df['gross']
avrage_grose_per_genre['gross'] = avrage_grose_per_genre_temp.groupby('genre')['gross'].mean()
avrage_grose_per_genre.sort_values('gross', ascending=False, inplace=True)
avrage_grose_per_genre
```

```python
# plt.rcParams['figure.figsize'] = [17, 10]
plt.bar(avrage_grose_per_genre.index, avrage_grose_per_genre['gross'])
plt.title("Gross Average on Genre")
plt.xlabel("Genre")
plt.ylabel("Gross Average")
```

## top director's having most gross films

```python
top_of = 6  # how to many top director per most gross films
sort_gross = df.sort_values('gross', ascending=False)
sort_gross
temp = {}
for index, row in sort_gross.iterrows():
    if row.director not in temp:
        temp[row.director] = index
    if len(temp) == top_of:
        break
temp
top_directos_most_gross_f = df.loc[temp.values()]
top_directos_most_gross_f
```
